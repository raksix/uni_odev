## Mantıksal Operatörler (Logical Operators) Örnekleri

```python
# Ve (and) operatörü ile kullanımlar
a = True
b = False

print(a and b)         # False
print(True and True)   # True
print(False and False) # False

# Veya (or) operatörü ile kullanımlar
print(a or b)         # True
print(True or False)  # True
print(False or False) # False

# Değil (not) operatörü ile kullanımlar
print(not a)   # False
print(not b)   # True
```

## Atama Operatörleri (Assignment Operators) Örnekleri

```python
x = 10   # x değişkenine 10 atanır
x += 3   # x'in değeri 3 arttırılır (10 + 3)
print(x) # 13

y = 20   # y değişkenine 20 atanır
y -= 4   # y'nin değeri 4 azaltılır (20 - 4)
print(y) # 16

z = 5    # z değerine 5 atanır
z *= 2   # z'nin değer 2 ile çarpılır (5*2)
print(z) # 10

a = 15   # a değişkenine 15 atanır
a /= 3   # a'nın değeri 3'e bolünür
print(a) # 5.0

b = 8    # b değişkenine 8 atanır
b **= 2  # b'nin değeri 2. üssüne alınır (8 ** 2)
print(b) # 64

c = 14   # c değişkenine 14 atanır
c %= 4   # c'nin değeri 4 ile mod alınır (14 % 4)
print(c) # 2

d = 20   # d değişkenine 20 atanır
d //= 3  # d'nin değeri 3'e tam sayı bölmesi yapılır (20 // 3)
print(d) # 6
```

## Karışık Basit Örnekler

## Karşılaştırma ve Mantıksal Operatörlerin Birlikte Kullanımı

```python
a = 10
b = 20
c = 15

# Karşılaştırma ve mantıksal işlemler
sonuc1 = a < b and b > c   # True and True -> True
sonuc2 = a == 10 or c > b  # True or False -> True
sonuc3 = not (a == b)      # not False     -> True

print("Sonuç 1:", sonuc1)  # True
print("Sonuç 2:", sonuc2)  # True
print("Sonuç 3:", sonuc3)  # True
```

## Atama Operatörleri ile Hesaplama

```python
x = 5
y = 10

x += 3    # x = x + 3  -> 8
x -= 2    # y = y - 2  -> 8
x *= 2    # x = x * 2  -> 16
y //= 4   # y = y // 4 -> 2

print("x:",x) # 16
print("y:",y) # 2
```

## Basit Bir Dönüşüm Hesaplama

## Basit Bir Hesaplama Programı

```python
# Ürün fiyatları
urun1 = 50
urun2 = 30
urun3 = 20

# Toplam tutar
toplam_tutar = urun1 + urun2 + urun3

# İndirim oranı (%10 indirim)
indirim_orani = 0.10

# İndirimli tutar
indirim_tutari = toplam_tutar * indirim_orani
odenecek_tutar = toplam_tutar - indirim_tutari

print("Toplam Tutar:", toplam_tutar)     # 100
print("İndirim Tutarı:", indirim_tutari) # 10.0
print("Ödenecek Tutar:", odenecek_tutar) # 90.0
```

## Mantıksal Operatörlerle Geçme Durumu

```python
# Öğrencinin sınav notları
vize = 60
final = 70

# Geçme notu
gecme_notu = 50

# Ortalama hesaplama
ortalama = (vize + final) / 2

# Mantıksal operatör kullanımı
geciyor_mu = ortalama >= gecme_notu
print("Ortalama Not:", ortalama)    # 65.0
print("Geçti mi?", geciyor_mu)      # True
```

## Aritmetik ve Atama Operatörleri ile Basit Maaş Artışı

```python
# Başlangıç maaşı
maas = 3000

# Zam oranı ve vergi oranı
zam_orani = 0.15
vergi_orani = 0.10

# Zam yapılır
maas += maas * zam_orani

# Vergi kesintisi yapılır
maas -= maas * vergi_orani

print("Yeni Maaş:", maas) # Zam ve vergi sonrası maaş
```

## Mantıksal Operatörlerle İndirimli Ürün Seçimi

```python
# Ürün bilgileri
fiyat        = 150
stok_miktarı = 5
yeni_urun_mu = True

# İndirim koşulları: Ürün 100 TL'den fazla, stok 3'te fazla ve yeni ürün ise
indirimli_mi = fiyat > 100 and stok_miktarı > 3 and yeni_urun_mu

print("Ürün İndirimli Mi?", indirimli_mi) # True
```

## Aritmetik Operatörlerle Basit Bir Yaş Farkı Hesaplama

```python
# Kişi yaşları
yas1 = 30
yas2 = 24

# Yaş farkı hesaplama
yas_farki = abs(yas1 - yas2)   # abs() fonksiyonu negatif sonucu pozitif yapar
print("Yaş Farkı:", yas_farki) # 6
```

## Mantıksal Operatörlerle Depo Kontrolü

```python
# Depo bilgileri
depo_kapasitesi = 1000
mevcut_malzeme = 850
yeni_girilecek_malzeme = 200

# Depo doluluğu kontrol
kullanilabilir_mi = (mevcut_malzeme + yeni_girilecek_malzeme) <= depo_kapasitesi
print("Depo Kullanılabilir Mi?", kullanilabilir_mi) # False
```

## Mantıksal ve Karşılaştırma Operatörleri ile Sınır Kontrolü

```python
# Sayı ve sınırlar
sayi = 15
alt_sinir = 10
ust_sinir = 20

# Sınır kontrolü
sinirda_mi = sayi >= alt_sinir and sayi <= ust_sinir
print("Sayı Sınırda mı?", sinirda_mi) # True
```

## Aritmetik Operatörlerle Mesafe Hesaplama

```python
# Noktaların koordinatları
x1,y1 = 3,4
x2,y2 = 7,1

# Mesafe formülü ile hesaplama
mesafe = ((x2 - x1) ** 2 + (y2 - y1) ** 2) ** 0.5
print("İki nokta arasındaki mesafe:", mesafe) # 5.0
```

## Mantıksal Operatörler ile Basit Yetki Kontrolü

```python
# Kullanıcı bilgileri
yonetici_mi = True
hesap_aktif = False

# Yetki kontrolü
yetkili_mi = yonetici_mi and hesap_aktif
print("Kullanıcı Yetkilisi Mi?", yetkili_mi) # False
```
